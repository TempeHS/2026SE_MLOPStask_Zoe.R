# Design

## Defining the business problem to be solved

When looking for exoplanets, trying to find the planets by themselves can be hard, as they do not have any noticeable features that would make them stand out, such as light emissions. However, they do block some of the light from stars when they orbit in front of them, and by observing the changes in a star's light intensity over time we can look for patterns indicating it is has one or more exoplanets orbiting it. 

## Refactoring the business problem into a machine learning problem
*First describe the problem in plain language, write down the problem in simple, direct words requires thinking through what really needs to be accomplished. It is important you identify what business outcome you are trying to achieve, not what process or model you want to use.*

**Students to refactor the provided business problem** in this markdown block

When given a set of data on the light fluctuations of a star, the machine learning model will classify the star as either having one or more exoplanets or having none, based on the patterns found in the training data. 

## Defining success metrics

*We are not talking about MSE, MAE, or R2 score. The success metrics should be focused on business, addressing the ideal outcome you identified earlier. How will you determine if the machine learning model fails to live up to business requirements? How will you know it has succeeded?*


**Students to define success metrics** in this markdown block

A successful model will predict whether or not a star has any exoplanets orbiting it with the same or a greater level of accuracy than human predictions, making it easier for scientists to locate new exoplanets.
