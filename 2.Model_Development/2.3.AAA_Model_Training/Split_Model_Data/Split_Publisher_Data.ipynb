Splitting the data based on publisher. 

```python
import pandas as pd
dataframe = pd.read_csv('model_ready_data.csv')
```

```python
AAA_dataframe = dataframe[dataframe['Publisher'] == 1]
AA_dataframe = dataframe[dataframe['Publisher'] == 2]
Indie_dataframe = dataframe[dataframe['Publisher'] == 3]
```

```python
AAA_dataframe.to_csv('model_ready_AAA.csv', index=False)
AA_dataframe.to_csv('model_ready_AA.csv', index=False)
Indie_dataframe.to_csv('model_ready_Indie.csv', index=False)
```
