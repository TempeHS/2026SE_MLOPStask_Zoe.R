## Defining the business problem to be solved

In the video game industry, there are numerous factors which go into how well a game sells, which emerging (or even pre-existing developers) may want to examine in order to see how well their game might sell. By looking at how much these factors have influenced the sales of other games, developers can use this to determine factors about their own - such as what console they release it on, who publishes it, what genre they make it, etc. 

## Refactoring the business problem into a machine learning problem
*First describe the problem in plain language, write down the problem in simple, direct words requires thinking through what really needs to be accomplished. It is important you identify what business outcome you are trying to achieve, not what process or model you want to use.*

I intend to create a machine learning model which, when given specific data about a video game, compares this to the data of other video games, and using this predicts how well a game based on the inputs provided would sell. 

## Defining success metrics

*We are not talking about MSE, MAE, or R2 score. The success metrics should be focused on business, addressing the ideal outcome you identified earlier. How will you determine if the machine learning model fails to live up to business requirements? How will you know it has succeeded?*


**Students to define success metrics** in this markdown block

A successful model will give mostly accurate predictions on how well a game would do, based on certain factors such as genre and publisher. It is not intended to give a fully accurate prediction, as the data is far too vast for the prediction to be exact, and a machine learning model can never fully predict the circumstances following a game's release, however the model should be able to give a general idea for the sales following a game's release.
