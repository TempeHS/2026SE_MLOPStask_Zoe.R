```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_absolute_error,mean_squared_error 
```

```python

training = pd.read_csv('training_data.csv')

x_name = [['Platform', 'Genre', 'Publisher']]
y_name = ['Global_Sales']

training.head()
```

```python

ax = plt.scatter(training.Platform, training.Genre, c=training['Publisher'])
plt.colorbar()
plt.show()
```

```python
regression = LinearRegression()

poly = PolynomialFeatures(degree=2, include_bias=False)

poly_features = poly.fit_transform(x_name)

regression.fit(poly_features, y_name)
```
